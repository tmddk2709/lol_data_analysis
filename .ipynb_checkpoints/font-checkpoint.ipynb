```python
matches_df = pd.read_csv('./data/lucky_score_matches_cleansed.csv')
matches_df.drop(columns=['Unnamed: 0'], inplace=True)
matches_df.tail()
```

```python
for c in matches_df.columns:
    print(c)
```

```python
matches_df[['win', 'firstBlood', 'summonerName']].groupby(by=['summonerName', 'win']).sum().plot.bar(figsize=(15,7))
```

```python
matches_df[['win', 'firstBlood', 'summonerName']].groupby(by=['win', 'summonerName']).sum().plot.bar(figsize=(15,7))
```

```python
matches_df[['win', 'firstBlood', 'summonerName']].groupby(by=['summonerName']).sum().plot.bar(figsize=(15,7))
```

```python
matches_df['firstBlood'] = matches_df['firstBlood'].apply(lambda x: 1 if x==True else 0)

win = matches_df.loc[matches_df['win'] == True]
# win['firstBlood'] = win['firstBlood'].apply(lambda x: 1 if x==True else 0)

lose = matches_df.loc[matches_df['win'] == False]
# lose['firstBlood'] = lose['firstBlood'].apply(lambda x: 1 if x==True else 0)

matches_gb = matches_df.groupby(by='summonerName').mean()
win_gb = win.groupby(by='summonerName').mean()
lose_gb = lose.groupby(by='summonerName').mean()
```

```python
plt.figure(figsize=(15,7))
p1 = plt.bar(matches_gb.index, matches_gb['firstBlood'], color='tab:blue')
plt.xticks(matches_gb.index, rotation=45)
plt.suptitle('본인 팀 퍼블 확률', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(15,7))
p1 = plt.bar(win_gb.index, win_gb['firstBlood'], color='tab:blue')
p2 = plt.bar(lose_gb.index, lose_gb['firstBlood'], color='tab:orange')
plt.xticks(win_gb.index, rotation=45)
plt.suptitle('승리 시 본인 팀 퍼블 확률', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(15,7))
p1 = plt.bar(lose_gb.index, lose_gb['firstBlood'], color='tab:blue')
plt.xticks(lose_gb.index, rotation=45)
plt.suptitle('패배 시 본인 팀 퍼블 확률', fontsize=15)
plt.show()
```
