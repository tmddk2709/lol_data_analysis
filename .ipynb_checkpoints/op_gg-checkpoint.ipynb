```python
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup
import lxml.html
import re
```

```python
url = 'https://www.op.gg/summoner/champions/userName=hide+on+bush'
```

```python
html = urlopen(url)
bsObject = BeautifulSoup(html, 'html.parser')
# select = soup.head.find_all('div')

# for div in select:
#     print(div.get('content'))

for div in bsObject.find_all('table'):
    print(div.get('class'))
# print(bsObject)
```

```python
columns = bsObject.select('div > table > thead > tr > th')
columnlist = []

for col in columns:
    columnlist.append(col.text)
    
df = pd.DataFrame(columns=columnlist)

contents = bsObject.select('div > table > tbody > tr')

regex = re.compile(r'[\d\w.,:& ]+')

dfcontent = []

for con in contents:
    tds = con.find_all('td')
    del tds[1]
    
    for td in tds:
        text = td.text
        texts = regex.findall(text)
        dfcontent.append(texts)
            
    ser = pd.Series(dfcontent, index=columnlist)
    df = df.append(ser, ignore_index=True)
    
    dfcontent = []
    
df.drop(columns=['#'], inplace=True)
df.tail(10)
```
