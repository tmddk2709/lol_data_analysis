1) 페이커 팀이 퍼블을 당할 확률

2) 페이커 맞라이너와의 골드차 vs 전체 팀 간의 골드 차

```python
url = 'https://kr.api.riotgames.com/'
api_key = 'RGAPI-092eff46-a388-4fc4-9d56-32685e663ca0'
api = '?api_key='+api_key
```

### 수집할 선수들 목록

```python
T1 = ['Hide on bush', 'T1 Roach', '칸 나', 'Aprox', 'T1 Cuzz', '엘 림', 'T1 Ellim',
      'T1 Closer', 'Teddy12345', 'T1 Gumayusi', '게복치숟가락', '롤오브레전드1234', 'T1 SoKuri']
DRX = ['모찌피치모찌피치', '윤소공원', 'DRX 홍창현', '요붕스1', 'DRX Deft', '3shotspop']
GenG = ['Gen G Rascal', 'Gen G Clid', '아진짜너무졸려', 'Gen G Ruler', '곽곽보성', 'Gen G Kellin', '댕청잇']
DAMWON = ['Lv 1 너구리', 'JUGKING', 'DWG Canyon', 'showmakerzzz', 'DWG ShowMaker',
         'DWG Ghost', 'DWG BeryL']
LPL_MID = ['Heart', '모두건강하게', 'knightzz']

player_list = T1 + DRX + GenG + DAMWON + LPL_MID 
```

### summoner data

```python
player_df = pd.DataFrame()

for player in player_list:
    req = requests.get(url + 'lol/summoner/v4/summoners/by-name/' + player + api)
    temp_df = json_normalize(req.json())
    player_df = player_df.append(temp_df, ignore_index=True)
    
player_df = player_df[['name', 'summonerLevel', 'id', 'accountId', 'puuid', 'revisionDate']]
player_df.set_index('name', inplace=True)
player_df.head()
```

### summoner match data

```python
#각 Player 별 match list
time.sleep(120)
player_df['match_list'] = ''

for index in player_df.index:
    ai = player_df.loc[index]['accountId']
    matches_req = requests.get(url + 'lol/match/v4/matchlists/by-account/' + ai + api)
    matches_df = json_normalize(matches_req.json()['matches'])
    
    #소환사의 협곡 솔랭만 골라내기
    match_list = matches_df.loc[matches_df['queue'] == 420]['gameId'].values
    player_df.at[index, 'match_list'] = match_list.tolist()
```

```python
player_df.head(3)
```

```python
# 각 Player 별 match 세부 데이터

matches_df = pd.DataFrame()

#number of players
n_player = len(player_df)
#player number
pn = 0
```

```python
rn = 0
while (pn < n_player):
    
    #player's summoner name
    player = player_df.iloc[pn].name
    #player's match list
    match_list = player_df.loc[player]['match_list']
    
    #number of matches
    n_match = len(match_list)
    #match number
    mn = 0
    
    while(mn < n_match):
        match_id = match_list[mn]
        
        #riot api는 2분에 100개의 요청만 accept 함
        try:            
            req = requests.get(url + 'lol/match/v4/matches/' + str(match_id) + api)
            match_df = json_normalize(req.json())
            match_df['summonerName'] = player
            
            teams_df = json_normalize(match_df['teams'][0])
    
            ptcp_df = json_normalize(match_df['participants'][0])
            ptcp_df.columns = ptcp_df.columns.str.strip().str.replace('stats.', '')
            ptcp_df.columns = ptcp_df.columns.str.strip().str.replace('timeline.', '')
            ptcp_df = ptcp_df.loc[:, ~ptcp_df.columns.duplicated()]
            participants_delete_list = ['killingSprees', 'doubleKills', 'tripleKills', 
                                       'quadraKills', 'pentaKills', 'combatPlayerScore', 'objectivePlayerScore', 
                                       'totalPlayerScore', 'totalScoreRank', 'role', 'lane']
            for i in range(10):
                participants_delete_list.append('playerScore'+str(i))
            ptcp_df = ptcp_df[ptcp_df.columns.drop(participants_delete_list)]
            ptcp_df = ptcp_df[ptcp_df.columns.drop(list(ptcp_df.filter(regex='Diff')))]

            ptcp_idt_df = json_normalize(match_df['participantIdentities'][0])
            ptcp_idt_df.columns = ptcp_idt_df.columns.str.strip().str.replace('player.', '')
            ptcp_idt_df = ptcp_idt_df[ptcp_idt_df.columns.drop(['platformId', 'currentPlatformId',
                                                                'matchHistoryUri', 'profileIcon'])]

            ptcp_df = pd.merge(ptcp_df, ptcp_idt_df, on='participantId')

            df = match_df[['gameId', 'gameCreation', 'gameDuration', 'queueId', 'mapId', 
                       'seasonId', 'gameVersion', 'gameMode', 'gameType', 'summonerName']]
            df = pd.merge(df, ptcp_df, on='summonerName', how='left')
            df = pd.merge(df, teams_df[['teamId', 'firstBlood']], on='teamId', how='left')

            matches_df = matches_df.append(df, ignore_index=True)
            mn = mn+1
            
        except Exception as e:
            if req.status_code == 405 or req.status_code == 403:
                print('{0}   {1} '.format(datetime.datetime.now().strftime("%H:%M:%S"), req.status_code))
                pass
            else:
                print('{0}   {1} '.format(datetime.datetime.now().strftime("%H:%M:%S"), e))
                time.sleep(5)
                continue
                
        rn = rn+1
        
        if (rn%100) == 0:
            print('{0}   {1:5d}: {2} {3}'.format(datetime.datetime.now().strftime("%H:%M:%S"), rn, player, match_id))
            matches_df.to_csv('./data/lucky_score_matches_cleansed.csv')
            time.sleep(120)
            
    pn = pn+1
```
